```python
# Import dependencies.
# Pyplot provides a MATLAB-like interface for making plots.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
# Return and unpack a tuple of figure & axes object(s)
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis) 
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
plt.xlim("Jan","Mar")
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Change the bar color to cyan.
# Add a legend for the city of Chicago.
# Add a title and axes labels.
# Switch the axis so that the Fare($) data is on the x-axis.
# Invert the y-axis data
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan",label="Chicago")
ax.legend()
# Create a title.
ax.set_title("PyBer Fare by Month :)")
# Create labels for the x and y axes.
ax.set_xlabel("Date :)")
ax.set_ylabel("Fare($) :)")
ax.invert_yaxis()
```

```python
# Create a Scatter Plot Using the Object-Oriented Interface

fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Create a Bubble Chart Using the Object-Oriented Interface
# Use list comprehension to size the data points as bubbles

fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[i*5 for i in y_axis], 
           c='skyblue', alpha=.5, linewidth=2, edgecolor='black'
          ,label='Boston')
ax.set_xlim(0,50)
ax.legend()
ax.set_title('PyBer Fare by Month')
ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')
ax.invert_yaxis()
```

```python
# Create a Pie Chart Using the MATLAB Method

plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
plt.subplots(figsize=(8,8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, 
        explode=explode_values, 
        labels=x_axis, 
        autopct='%.1f%%')
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
# Create a Pie Chart Using the Object-Oriented Interface

fig, ax = plt.subplots()
dir(ax)
explode_values = (0, 0, .2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

#ax.subplots(figsize=(8,8))
ax.pie(y_axis,
       labels=x_axis,
       autopct='%.1f%%',
       shadow=True,
       colors=colors,
       explode=explode_values,
       startangle=90,
       counterclock=False,
      radius=2)
plt.show()
```
