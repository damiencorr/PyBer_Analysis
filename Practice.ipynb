```python
# Import dependencies.
# Pyplot provides a MATLAB-like interface for making plots.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
# Return and unpack a tuple of figure & axes object(s)
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis) 
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
plt.xlim("Jan","Mar")
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# SKILLZ DRILL :)
# Change the line color to green.
# Change the marker to a diamond.
# Change the line width to 2 points.
# Add a legend for the city of Boston.
# Add a title and axes labels.
# Add a y-axis limit.
# Add grid lines.

# Create the plot.
plt.plot(x_axis, y_axis, marker="D", color="green", linewidth=20, label='Boston :)')
# Create labels for the x and y axes.
plt.xlabel("Date :)")
plt.ylabel("Fare($) :)")
# Set the y limit between 0 and 45.
plt.ylim(0, 49)
plt.xlim(0,14)
# Create a title.
plt.title("PyBer Fare by Month :)")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create the plot
plt.bar(x_axis, y_axis)
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create a horizontal chart using plt.barh():
# Create the plot
plt.barh(x_axis, y_axis)
```

```python
# Create a horizontal chart using plt.barh():
# Switch data to opposit axis
plt.barh(y_axis, x_axis)
```

```python
# Invert the y axis and have months in ascending order
# Create the plot.
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

```python
# Change the bar colors to magenta.
# Add a legend for the city of Boston.
# Add a title and axes labels.
# Invert the y-axis so that January is at the top.
plt.barh(x_axis, y_axis, color="magenta", label="Boston")
plt.gca().invert_yaxis()
plt.legend()
# Create a title.
plt.title("PyBer Fare by Month :)")
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
```

```python
# Create the plot using OBJECT ORIENTATION with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
ax.bar(x_axis, y_axis, color="black", label='New York')
ax.legend()
```

```python
# Creat horrizontal bar chart using OO interface
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
#ax.barh(y_axis, x_axis)
```

```python
# Change the bar color to cyan.
# Add a legend for the city of Chicago.
# Add a title and axes labels.
# Switch the axis so that the Fare($) data is on the x-axis.
# Invert the y-axis data
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan",label="Chicago")
ax.legend()
# Create a title.
ax.set_title("PyBer Fare by Month :)")
# Create labels for the x and y axes.
ax.set_xlabel("Date :)")
ax.set_ylabel("Fare($) :)")
ax.invert_yaxis()
```
